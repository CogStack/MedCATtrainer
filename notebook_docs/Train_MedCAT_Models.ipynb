```python
cdb_path = './data/cdb-medmen.dat'
vocab_path = './data/vocab.dat'
```

```python
cdb = CDB.load(cdb_path)
vocab = Vocab.load(vocab_path)
cat = CAT(cdb, vocab)
```

```python
with open("example_data/MedCAT_Export_With_Text_2020-05-22_10_34_09.json") as f:
    data = json.load(f)
cat.trainer.train_supervised_raw(
    data=data,
    nepochs=1,
    print_stats=True,
    use_filters=True)
```

```python
from medcat.components.addons.meta_cat import MetaCAT
from tokenizers import ByteLevelBPETokenizer
from itertools import chain
```

```python
import numpy as np
# Tokenizer instantiation
tokenizer = ByteLevelBPETokenizer(vocab_file='data/medmen-vocab.json', merges_file='data/medmen-merges.txt')
embeddings = np.load(open('data/embeddings.npy', 'rb'))
```

```python
metacat = MetaCAT(tokenizer=tokenizer, embeddings=embeddings,
             pad_id=len(embeddings) -1, save_dir='mc_status', device='cpu')
```

```python
proj = json.load(open('example_data/MedCAT_Export_With_Text_2020-05-22_10_34_09.json'))
```

```python
# Tasks we are training for...
set(chain.from_iterable([a['meta_anns'].keys() for d in proj['projects'][0]['documents'] for a in d['annotations']]))
```

```python
metacat.train('example_data/MedCAT_Export_With_Text_2020-05-22_10_34_09.json', 'Skip', nepochs=5)
```
